```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = plt.axes(projection='3d')

def draw_line(p0: tuple[int, int, int], p1: tuple[int, int, int]):
    x = [p0[0], p1[0]]
    y = [p0[1], p1[1]]
    z = [p0[2], p1[2]]
    ax.scatter(x, y, z)
    ax.plot(x, y, z)

a = np.array((0, 0, 0))
b = np.array((1, 1, 1))
dist = np.linalg.norm(a-b) 

draw_line(a, b)

mid = (b-a)/2
ax.text(mid[0] + .1, mid[1], mid[2], dist)
print(mid)
print(dist)

draw_line(a, (b[0], 0, 0))
draw_line(a, (0, b[1], 0))
draw_line(a, (0, 0, b[2]))


plt.show()
```
